```python
dfE=pd.read_excel('Értekesites(nagyker).xlsx','E')
dfE=dfE.pivot_table(index='Vevőnév',columns=['Kategórianév'],values='Nyereség',aggfunc='sum',margins=True)
dfE.rename(columns = {'All':'Nyereség'}, inplace = True)
dfE
```

```python
dfE=pd.read_excel('Értekesites(nagyker).xlsx','K1',skiprows=3)
dfE=pd.melt(dfE,id_vars=['Vevők','Végösszeg'],var_name="Város",value_name="Költség")
dfE.drop(dfE[dfE['Vevők'] == "Végösszeg"].index, inplace = True)
del dfE["Végösszeg"]
dfE
```
